```python
count=0
records = []

for qrel in dataset.qrels_iter():
    count +=1
#    print(qrel.doc_id , qrel.query_id, qrel.relevance)  
#    print(qrel.query_id,"####",queries.get(qrel.query_id),"####",qrel.doc_id,"####",docs.get(qrel.doc_id),"####" , qrel.relevance )
    # if count==10:
    #     break
    if(qrel.query_id=="121352"):
        print("found it")
    record = {
        "query_id": qrel.query_id,
        "query_text": queries.get(qrel.query_id),
        "doc_id": qrel.doc_id,
        "doc_text": docs.get(qrel.doc_id),
        "relevance": qrel.relevance
    }
    records.append(record)
   

print(count)

```

```python

def get_by_query_id(records, qid):
    return [r for r in records if r["query_id"] == qid]

print(get_by_query_id(records, "737889"))
```
